```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

***EJERCICIO UNO***

```python
# Cargar datos
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columnas = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 
            'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
data = pd.read_csv(url, names=columnas)
print(f"Datos cargados: {data.shape[0]} filas y {data.shape[1]} columnas")
print(data.head())
```

```python
# Separar características y objetivo
X = data.drop('Outcome', axis=1)
y = data['Outcome']

# Dividir datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar modelo
from sklearn.linear_model import LogisticRegression
modelo = LogisticRegression(max_iter=1000)
modelo.fit(X_train, y_train)

```

```python
# Predecir y evaluar
y_pred = modelo.predict(X_test)
print("Exactitud:", accuracy_score(y_test, y_pred))
print("\nMatriz de confusión:\n", confusion_matrix(y_test, y_pred))
print("\nReporte de clasificación:\n", classification_report(y_test, y_pred))
```

```python
# Cargar datos
url = "https://raw.githubusercontent.com/arib168/data/main/Social_Network_Ads.csv"
data = pd.read_csv(url)

print(f"Datos cargados: {data.shape[0]} filas y {data.shape[1]} columnas")
print(data.head(20))
```

```python
compraron = data[data['Purchased'] == 1]
print(compraron.head(20))  # Mostrar las primeras filas de los que compraron
```

```python
edadMayor = data[data['Age'] <= 22]
print(edadMayor.head(20))
```

```python
# Seleccionar características y objetivo
X = data[['Age', 'EstimatedSalary']]
y = data['Purchased']

# Estandarizar características
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Dividir datos
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Crear y entrenar modelo
from sklearn.linear_model import LogisticRegression
modelo = LogisticRegression()
modelo.fit(X_train, y_train)
```

```python
# Evaluar modelo
y_pred = modelo.predict(X_test)
print("Exactitud:", accuracy_score(y_test, y_pred))
print("\nReporte de clasificación:\n", classification_report(y_test, y_pred))
```

```python
# Función para predecir compra con nuevos datos
def predecir_compra(edad, salario):
    # Crear dataframe con los datos nuevos
    datos_nuevos = pd.DataFrame([[edad, salario]], columns=['Age', 'EstimatedSalary'])
    
    # Escalar datos usando el scaler original
    datos_nuevos_scaled = scaler.transform(datos_nuevos)
    
    # Predecir con el modelo
    prediccion = modelo.predict(datos_nuevos_scaled)
    probabilidad = modelo.predict_proba(datos_nuevos_scaled)[0][1]  # probabilidad de comprar (clase 1)
    
    if prediccion[0] == 1:
        print(f"Predicción: La persona probablemente COMPRARÁ el producto (Probabilidad: {probabilidad:.2f})")
    else:
        print(f"Predicción: La persona probablemente NO COMPRARÁ el producto (Probabilidad: {probabilidad:.2f})")

# Ejemplo de uso:
edad_input = float(input("Ingresa la edad: "))
salario_input = float(input("Ingresa el salario estimado: "))

print(f"Salario estimado: {salario_input}")
print(f"Edad: {edad_input}")
predecir_compra(edad_input, salario_input)
```
