```python
import pandas as pd

datos = pd.read_csv("./data/celsius.csv")
```

```python
X = datos["Celsius"].values
y = datos["fahrenheit"].values
```

```python
import tensorflow as tf

modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape = (1,)),
    tf.keras.layers.Dense(units = 1),
])

modelo.compile(
    optimizer = tf.keras.optimizers.Adam(.1),
    loss = "mean_squared_error"
)
```

```python
entrenamiento = modelo.fit(X, y, epochs = 1000)
```

```python
import matplotlib.pyplot as plt

plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(entrenamiento.history["loss"])
plt.show()
```

```python
import numpy as np
c = int(input("Ingrese un numero en celcius: "))
prediccion = modelo.predict(np.array([c]))

print(f"Prediccion: {c} celsius son {prediccion[0]} fahrenheit")
print(f"Prediccion: {c} celsius son {prediccion[0]} fahrenheit")
```
